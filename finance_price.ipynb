## 예측 모델이 얼마나 잘 맞는지 종목별 테스트(시계열 테스트 X)

```python
import pandas as pd
import numpy as np
import math as m
import datetime
import os
import dgroot as dg


# 종목코드 받아오기
code_df = pd.read_html('http://kind.krx.co.kr/corpgeneral/corpList.do?method=download&searchType=13', header=0)[0]
code_df.종목코드 = code_df.종목코드.map('{:06d}'.format)
code_df = code_df[['회사명', '종목코드']]
code_df = code_df.rename(columns={'회사명': 'name', '종목코드': 'code'})
```

### 데이터 받아오는 함수 정리

```python
def get_name(code_df_):
    name_list = code_df_['name'].values.tolist()
    return name_list

def get_url(item_name_, code_df_):
    code = code_df_.query("name=='{}'".format(item_name_))['code'].to_string(index=False)
    url_ = 'http://finance.naver.com/item/sise_day.nhn?code={code}'.format(code=code.strip())

    print("요청 URL = {}".format(url_))
    return url_

def get_data(url, gubun, page_amount):
    df = pd.DataFrame()
    for page in range(1, page_amount + 1):
        pg_url = '{url}&page={page}'.format(url=url, page=page)
        df = df.append(pd.read_html(pg_url, header=0)[0], ignore_index=True)
    df = df.dropna()
    if gubun == 1:
        df = df.rename(columns={'날짜': 'date', '종가': 'close', '전일비': 'diff', '시가': 'open', '고가': 'high', '저가': 'low',
                                    '거래량': 'volume'})
        df[['close', 'diff', 'open', 'high', 'low', 'volume']] = df[['close', 'diff', 'open', 'high', 'low', 'volume']].astype(int)
    else:
        print(df)
        df = df.rename(columns={'날짜': 'date', '체결가': 'close', '전일비': 'diff', '등락률': 'percent', '거래량': 'volume', '거래대금': 'amount'})
        df[['close', 'diff', 'volume', 'amount']] = df[['close', 'diff', 'volume', 'amount']].astype(int)
    # df['date'] = pd.to_datetime(df['date'])
    date_ = df['date'].values.tolist()
    price_ = df['close'].values.tolist()
    return [date_, price_]
```

### 분석모델

```python
def check_vision(price):
    # 1차 데이터(재료)
    diff = dg.difference(price, 1)
    shape = dg.difference(diff, 0)
    shape_net = dg.sum_partial(shape, 0, 5)
    stand = dg.sum_data(price, shape, 1, 0)
    price_stand = dg.sub_data(price, stand, 0, 0)

    # 2차 데이터(부호화)
    diff = dg.sign(diff, 0)
    shape = dg.sign(shape, 0)
    shape_net = dg.sign(shape_net, 0)
    price_stand = dg.sign(price_stand, 0)

    # 3차 데이터(합손질)
    diff_shape_sum = dg.sum_data(diff, shape, 0, 0)
    shape_shape_net_sum = dg.sum_data(shape, shape_net, 0, 0)
    shape_sum = dg.sum_partial(shape, 0, 2)
    shape_net_sum = dg.sum_partial(shape_net, 0, 2)
    price_stand_sum = dg.sum_partial(price_stand, 0, 2)

    # 3차 데이터(유량 손질)
    shape_flow = dg.sum_partial(shape, 0, 5)
    shape_net_flow = dg.sum_partial(shape_net, 0, 5)
    price_stand_flow = dg.sum_partial(price_stand, 0, 5)

    shape_flow = dg.sub_partial(shape_flow, 0, 1)
    shape_net_flow = dg.sub_partial(shape_net_flow, 0, 1)
    price_stand_flow = dg.sub_partial(price_stand_flow, 0, 1)
    slope_ = float(sum(diff[1:23]) / 22)

    result = []
    checking = []
    for i in range(1, min(len(shape_shape_net_sum), len(shape_net_flow))):
        res = -(diff_shape_sum[i] + price_stand_sum[i] + shape_net_sum[i] + shape_sum[i] + shape_shape_net_sum[i] + price_stand_flow[i] + shape_flow[i] + shape_net_flow[i])
        checking.append(res)
        if res == 0:
            result.append(0)
        else:
            result.append(abs(res)/res)

    # print(checking[0])

    return {'result': result, 'difference': diff, 'slope': slope_}


def new_check_vision(price, amount):
    choice_data_num = amount // 5
    # 1차 데이터(재료)
    diff = dg.difference(price, 1)
    shape = dg.difference(diff, 0)
    shape_diff = dg.difference(shape, 1)

    # 2차 데이터
    shape_diff_sign = dg.sign(shape_diff, 0)
    shape_diff_abs = dg.abs_data(shape_diff, 0)
    slope = dg.sum_partial(diff, 0, 5)
    slope_ = float(sum(diff[1:23])/22)

    # 3차 데이터
    shape_diff_average = np.average(np.array(shape_diff[2:]))
    choice_shape_diff_av = float(sum(shape_diff[2:(2 + choice_data_num)])) / float(choice_data_num)
    shape_diff_standard = float(np.std(np.array(shape_diff[2:])))
    if shape_diff_standard == 0.0:
        shape_diff_standard = 0.01
    weight = m.exp(float(pow(shape_diff_average, 2) - pow(choice_shape_diff_av, 2))/(2.0*pow(shape_diff_standard, 2)))
    predict_average = choice_shape_diff_av * weight
    choice_shape_diff_sum = sum(shape_diff[2:(1 + choice_data_num)])
    predict_data = predict_average * choice_data_num - choice_shape_diff_sum

    predict_data += shape[1] + diff[1]

    return [predict_data, slope_]
```

### vision 분석 모델을 이용하여 횡으로 분석하기

```python
name_list = get_name(code_df)

url = get_url('삼성전자', code_df)
buffer = get_data(url, 1, 1)
date = buffer[0]

ob_initial_day = int(input("관측 시작 일자를 현재 기준으로 몇 일전으로 하시겠습니까?: "))
ob_day = int(input('관측 기간을 몇 일간으로 하시겠습니까?: '))
expect_day = int(input('예측 범위 일자를 몇 일로 하시겠습니까?: '))

total_accuracy = 0.0
for ob in range(ob_initial_day, ob_initial_day + ob_day):
    accuracy = []
    for item_name in name_list:
        path = 'C:/VISION_FINAL_0.4/종목가격데이터/'
        title = str(date[0]) + ';' + str(item_name) +  '.csv'
    
        try:
            f = open(path + title, 'rt')
            f.close()
            pass
        except:
            continue
    
        data = pd.read_csv(path + title).values.tolist()[:ob_day + ob_initial_day + 30]
        price = [data[i][1] for i in range(len(data))]
        buffer = check_vision(price)
        result = buffer['result']
        diff = buffer['difference']
        slope = buffer['slope']
    
        check = 0
        try:
            for i in range(1, expect_day + 1):
                buffer = price[ob] - price[ob + i]
                sign = 0
                if buffer > 0:
                    sign = 1
                elif buffer < 0:
                    sign = -1
                else:
                    sign = 0
        
                if sign == result[ob + 1]:
                    check = 1
        except IndexError:
            continue
    
        if check == 1:
            accuracy.append(1)
        else:
            accuracy.append(0)

    accuracy = float(sum(accuracy) / len(accuracy))
    total_accuracy += accuracy

    print(str(ob) + ' 일 전부터 ' + str(expect_day) + '일간 횡 정확도는 ' + str(accuracy) + ' (%) 입니다.')

total_accuracy /= ob_day

print('\n전체 관측 기간 동안의 정확도는 ' + str(total_accuracy) + '입니다.')
```
